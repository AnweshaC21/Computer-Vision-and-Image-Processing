Reading an image using openCV function, displaying by plotting the image using matplotlib library, saving it and writing a piece of text such as a name using pixel operations.

```python
# importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# read the image (from google drive)
from google.colab import drive
drive.mount('/content/gdrive')

img = cv2.imread("/content/gdrive/MyDrive/Datasets/greyscaleimage1.jpg")
```

A greyscale image is read from drive into 'img'.

```python
# display the image
plt.imshow(img)
plt.show()
```

The greyscale image is plotted on the xy-plane.

```python
# save the image
cv2.imwrite('window.jpg', img)
```

The image is saved and can be downloaded as a file.

```python
# write name on the image by combining another image having the text on it

img.shape
```

```python
# A blank image with name written is created using paint, having the same dimensions as the previous image, read and plotted
name = cv2.imread("/content/gdrive/MyDrive/Datasets/name.jpg")
name.shape
```

```python
plt.imshow(name)
plt.show()
```

```python
# creating new empty image
#new_img = np.zeros((437,660,3))

# changing pixel values of 'img' according to 'name'
for i in range(437):
  for j in range(660):
    for k in range(3):
      if name[i,j,k] < 200:
        img[i,j,k] = 255
```

```python
# plotting the final image with name written
plt.imshow(img)
plt.show()
```

To write the name on the image 'img', first the image 'name' is scanned for pixel values less than 200, i.e. the non-white pixels, and the corresponding pixel values in 'img' are replaced with the value 255 to write the text in white colour.
